### Merging multiple csvs

```python
df = pd.DataFrame()
for file in glob.glob("*ig.csv"):
    brand = file.split("_ig.csv")[0]
    temp_df = pd.read_csv(file)
    temp_df["brand"] = brand
    df = pd.concat([df,temp_df], axis=0)
```

```python
def handle_likes(like):
    like = str(like).replace(",","")
    if "k" in like:
        return int(like.replace("k","000").replace(".",""))
    if "m" in like:
        return int(like.replace("m","000000").replace(".",""))
    return int(like)
        
```

```python
df = df[df.media=="image"]
```

```python
df.likes = df.likes.apply(handle_likes)
```

```python
df = df[df.caption.notnull()]
```

```python
y = df.likes
```

```python
X = df.caption
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(stop_words='english', max_df=0.3, min_df=4)
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
X_train_dtm = vect.fit_transform(X_train)
X_test_dtm = vect.transform(X_test)
```

```python
X_train_dtm.shape
```

```python
import sklearn.linear_model
import sklearn.datasets
import sklearn.svm
from sklearn.metrics import r2_score
import sklearn.decomposition
import sklearn.feature_extraction.text
import sklearn.utils.sparsefuncs
```

```python
pca = sklearn.decomposition.TruncatedSVD(n_components=100)
X_train_red = pca.fit_transform(X_train_dtm)
X_test_red = pca.transform(X_test_dtm)
```

```python
X_train_red.shape
```

```python
regression = sklearn.linear_model.LinearRegression()
regression.fit(X_train_red, y_train)
```

```python
yhat = regression.predict(X_test_red)
```

```python
r2_score(y_test,yhat)
```

```python
import numpy as np

def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true) + 1, np.array(y_pred) + 1
    return np.mean(np.abs((y_true - y_pred)/ y_true)) * 100
```
